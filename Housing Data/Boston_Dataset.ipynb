# <span style="color:red">Why do we need to perform feature scaling?
Scaling the features makes the flow of gradient descent smooth and helps algorithms quickly reach the minima of the cost function. Without scaling features, the algorithm may be biased toward the feature which has values higher in magnitude.

# <span style="color:green">Feature Scaling

```python

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_train_scaled
```

```python
X_test_scaled=scaler.transform(X_test)
X_test_scaled
```

# <span style="color:Green">Model Training


```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression
```

```python
regression.fit(X_train_scaled,y_train)
```

### <span style="color:blue"> Print the coefficients and intercept

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## Prediction for the test data
y_pred=regression.predict(X_test_scaled)
y_pred
```

# <span style="color:Red"> Assumption of linear regression

  * Linearity: The relationship between X and the mean of Y is linear.
  * Homoscedasticity: The variance of residual is the same for any value of X.
  * Independence: Observations are independent of each other.
  * Normality: For any fixed value of X, Y is normally distributed

# <span style="color:green">Showing assumption of linear regression

```python
plt.scatter(y_test,y_pred)
plt.xlabel("Test truth data",color='red')
plt.ylabel("Test predicted dta",color='blue')
```

### <span style="color:blue">Residuals

```python

residuals=y_test-y_pred
residuals
```

```python
sns.displot(residuals,kind="kde")
```

### <span style="color:blue">Scatter plot with predictions and residual

```python
##uniform distribution
plt.scatter(y_pred,residuals)
plt.xlabel("Uniform Distribution",color="red")
```

### <span style="color:blue">Performance Metrics for Linear Regression

```python

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

### <span style="color:blue">R2-score for Linear Regresssion


```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print(score)
```

### <span style="color:blue">Adjusted R2-score for Linear Regresssion


```python

1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

# <span style="color:Green">Ridge Regression Algorithm
